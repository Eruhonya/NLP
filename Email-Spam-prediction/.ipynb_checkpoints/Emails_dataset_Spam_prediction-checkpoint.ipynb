# Email Dataset

## Dataset Description

### Enron Email Dataset

The **Enron Email Dataset** emerged from the investigation into one of the largest corporate scandals in U.S. history — the collapse of the energy giant Enron Corporation in December 2001.

#### Enron's Background

* **Enron was one of the largest energy companies in the United States.**
* In 2001, the company went bankrupt due to massive accounting fraud and financial manipulation (the Enron scandal).
* This collapse led to a widespread criminal investigation.

#### Subpoena and Seizure of Emails

* During the investigation, the U.S. Department of Justice (DOJ) seized a massive volume of internal emails from the company's employees.
* The correspondence primarily took place among executives, managers, and senior staff.

#### Data Publication (2003–2004)

* In 2003, the Federal Energy Regulatory Commission (FERC) made a significant portion of this correspondence publicly available as part of its investigation.
* In 2004, researchers at the University of California, Berkeley, and other institutions cleaned, structured, and formatted the raw emails, making the dataset suitable for analysis and machine learning research.

### Use Cases

The **Enron Dataset** remains one of the very few real-world, publicly available large-scale corporate email corpora. It has become the gold standard for:

* **Natural Language Processing (NLP) tasks** (such as topic modeling, text summarization, and named entity recognition).
* **Social Network Analysis (SNA)** and communication flow mapping.
* **Machine Learning applications**, including spam detection, sentiment analysis, and anomaly detection.
* **Organizational behavior studies**, analyzing corporate culture, hierarchy, and internal communication patterns.

### Dataset Contains:

* **Over 500,000 emails.**
* **Around 150 users** (predominantly senior Enron employees).
* **Structure:** Folders such as "inbox", "sent", "deleted items", and others (reflecting the original email client structure).
* **Storage Format:** Commonly distributed as `.mbox` files or as a nested directory tree of text files.

### Variables / Features:

* **`text`** — The raw text content of the email.
* **`spam`** — A binary indicator showing whether the email is spam ($1$) or ham ($0$).

# Step 1 import libraries

```python
import numpy as np  # Arrays and mathematics
import pandas as pd  # DataFrames and data manipulation
import matplotlib.pyplot as plt  # Core data visualization
import seaborn as sns  # Advanced statistical plots
import string  # Built-in string punctuation utilities
import re  # Regular expressions for text cleaning
import time  # Execution time tracking
import timeit  # Precise code benchmarking
from collections import Counter  # Word frequency counting
from dateutil import parser
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS  # NLP stop words filtering
from sklearn.model_selection import train_test_split  # Dataset splitting (train/test)
from sklearn.pipeline import Pipeline  # ML pipeline workflow chaining
from sklearn.feature_extraction.text import TfidfVectorizer  # Text-to-vector conversion (TF-IDF)
from sklearn.svm import SVC  # Support Vector Classifier model
from sklearn.ensemble import RandomForestClassifier  # Random Forest Classifier model
from xgboost import XGBClassifier  # Gradient Boosting Classifier model
from sklearn.linear_model import LogisticRegression  # Logistic Regression model
from sklearn.naive_bayes import MultinomialNB  # Naive Bayes for text classification
from sklearn.metrics import (  # Model evaluation metrics:
    accuracy_score,  # - Overall accuracy classification score
    precision_score,  # - Precision metric (positives correctness)
    recall_score,  # - Recall metric (true positives capture rate)
    f1_score,  # - F1-score (harmonic mean of precision & recall)
    roc_auc_score,  # - ROC-AUC score for class separation quality
    confusion_matrix,  # - Raw confusion matrix array
    ConfusionMatrixDisplay,  # - Confusion matrix visual plotting tool
    classification_report  # - Summary text report of all main metrics
)
```

```python
def extract_date(text):
    try:
        first_line = text.splitlines()[0]  
        return parser.parse(first_line, fuzzy=True) 
    except:
        return pd.NaT

def evaluate_and_plot_pipelines(pipelines, X_train_data, X_test_data, y_train, y_test):
    for model_name, model_pipeline in pipelines.items():
        print(f"\n" + "="*50)
        print(f" Statistic for model: {model_name.upper()} ")
        print("="*50)
        
        for dataset_name, (X_data, y_true) in [('Train', (X_train_data, y_train)), ('Test', (X_test_data, y_test))]:
            
            y_pred = model_pipeline.predict(X_data)
            
            cm = confusion_matrix(y_true, y_pred, labels=[0, 1])
            
            fig, ax = plt.subplots(figsize=(5, 5))
            disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Ham', 'Spam'])
            disp.plot(cmap='Blues', colorbar=False, ax=ax)
            plt.title(f'Confusion Matrix — {model_name} ({dataset_name} Set)', fontsize=12, pad=10)
            plt.show()
            
            print(f'--- Classification Report — {model_name} ({dataset_name} Set) ---')
            print(classification_report(
                y_true, y_pred,
                labels=[0, 1],
                target_names=['Ham', 'Spam'],
                zero_division=0  
            ))
            print("\n" + "."*50 + "\n")

def plot_train_test_metrics(df_metrics, metrics=None):

    if metrics is None:
        metrics = ['Accuracy', 'Precision', 'Recall', 'F1-score', 'ROC-AUC']
        
    for metric in metrics:

        if metric not in df_metrics.columns:
            continue
            
        pivot_df = df_metrics.pivot(
            index='Model', columns='Dataset', values=metric
        )[['Train', 'Test']]

        ax = pivot_df.plot(kind='bar', figsize=(12, 6))
        plt.title(f'{metric} — Train vs Test', fontsize=14, pad=15)
        plt.ylabel(metric, fontsize=12)
        plt.xlabel('Модель', fontsize=12)
        plt.xticks(rotation=0)
        plt.ylim(0, 1.1)
        plt.grid(axis='y', linestyle='--', alpha=0.5) 

        for p in ax.patches:
            v = p.get_height()
            if not np.isnan(v):
                ax.annotate(
                    f'{v:.1%}',
                    (p.get_x() + p.get_width() / 2, v),
                    ha='center', va='bottom', fontsize=9.5, fontweight='bold'
                )

        plt.tight_layout()
        plt.show()
```

## Research Note

### 1. Chronological Noise (Artifacts from Past and Future)
Initial exploratory data analysis (EDA) revealed extreme timestamp anomalies, with email dates stretching into the past (1980, 1981, 1986) and the future (May/June 2026).  
This is a classic parsing artifact generated by `dateutil.parser(fuzzy=True)`. When processing unstructured spam emails that lack formal corporate headers, the parser aggressively scans the `Subject:` line for any numeric values. Random numbers—such as a "80% discount" or transaction IDs like "05" and "06"—were mistakenly converted into years. Furthermore, when a specific year was missing from the text, the library automatically backfilled the empty feature with the current system year (2026).  
**Resolution:** Since Enron Corporation was formed in 1985 and dissolved following its bankruptcy in late 2001, a strict chronological mask was applied to restrict the dataset to 1998–2002. This completely eliminated the algorithmic noise while preserving the actual historical window.

### 2. Post-Bankruptcy Reality (The 2002 Emails)
Despite Enron officially declaring bankruptcy in December 2001, the presence of valid corporate communication (`ham`) throughout early 2002 is historically accurate. According to Carnegie Mellon University documentation, the official Enron Corpus covers data up until early 2002.  
*Reality*: The company's liquidation, asset distribution, internal legal audits, and federal criminal investigations lasted for years. Key legal, IT, and administrative departments remained operational in early 2002 to facilitate the corporate wind-down and cooperate with the Department of Justice (DOJ).

### 3. Outliers
* Ultra-Short Emails (2–20 words): Manual inspection showed that very brief texts (e.g., `Subject: fyi`) are legitimate corporate quick-replies or document forwards. They represent genuine operational behavior rather than corrupted rows, meaning the future model must learn to extract intent from minimal textual features without misclassifying them as noise.
* Ultra-Long Emails (8,000+ words): The absolute longest messages in the dataset were verified as massive media briefings and news clips from the Indian press regarding the Enron Dabhol Power Company crisis in 2001. 

### 4. Feature Engineering
These EDA discoveries provide critical guidelines for the upcoming preprocessing pipeline:
1. **Header Stripping:** Technical prefixes like `Subject:` must be removed to avoid heavy feature bias.
2. **Noise Reduction:** Isolate and remove raw numbers, punctuation, and URLs (prevalent in long news digests) using regular expressions (`re`) to prevent the `TfidfVectorizer` from bloating the feature space with non-semantic tokens.

# Step 4 Feature engineering and model building

```python
text_series = df['text'].astype(str)

df['email_length'] = text_series.str.len()
df['num_words'] = text_series.str.split().str.len()
df['uppercase_ratio'] = text_series.str.findall(r'[A-Z]').str.len() / (df['email_length'] + 1e-5)
df['num_exclamations'] = text_series.str.count('!')

spammy_words = ['free', 'win', 'money', 'urgent', 'discount', 'offer', 'click', 'buy', 'cash', 'congratulations']
df['has_spammy_words'] = text_series.str.lower().str.contains('|'.join(spammy_words)).astype(int)


# Cleaning text and performing frequency analysis
cleaned_text = (
    text_series
    .str.replace(r'^Subject:\s*', '', case=False, regex=True)
    .str.replace(r'http\S+|www\S+', '', regex=True)
    .str.lower()
    .str.replace(r'[^\w\s]', '', regex=True)
) # Remove 'Subject:', links, punctuation, and convert to lowercase


all_words = " ".join(cleaned_text).split()
```

```python
print("\nTop 10 Most Frequent Words (Raw):")
print(Counter(all_words).most_common(10))

custom_stop_words = set(ENGLISH_STOP_WORDS).union({'enron'})
filtered_words = [w for w in all_words if w not in custom_stop_words and len(w) > 1]

print("\nTop 10 Most Frequent Semantic Tokens (Cleaned):")
print(Counter(filtered_words).most_common(10))
```

## Pipeline with metadata features

```python
features = ['text', 'email_length', 'num_words', 'uppercase_ratio', 'num_exclamations', 'has_spammy_words']
X = df[features]
y = df['spam'].astype(int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
```

```python
print(len(X_train), len(y))  
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
```

```python
# Preprocessors
meta_features = ['email_length', 'num_words', 'uppercase_ratio', 'num_exclamations', 'has_spammy_words']
preprocessor_svm = ColumnTransformer([
    ('tfidf', TfidfVectorizer(stop_words='english', ngram_range=(1, 2), min_df=3, max_df=0.95, max_features=5000), 'text'),
    ('meta', StandardScaler(), meta_features)
])
preprocessor_rf = ColumnTransformer([
    ('tfidf', TfidfVectorizer(stop_words='english', max_features=10000, ngram_range=(1, 2), min_df=5, max_df=0.9), 'text'),
    ('meta', StandardScaler(), meta_features)
])
preprocessor_xgb = ColumnTransformer([
    ('tfidf', TfidfVectorizer(stop_words='english', max_features=3000, ngram_range=(1, 2), min_df=3, max_df=0.95), 'text'),
    ('meta', StandardScaler(), meta_features)
])
preprocessor_lr = ColumnTransformer([
    ('tfidf', TfidfVectorizer(stop_words='english', ngram_range=(1, 2), min_df=3, max_df=0.95, max_features=5000), 'text'),
    ('meta', StandardScaler(), meta_features)
])
preprocessor_nb = ColumnTransformer([
    ('tfidf', TfidfVectorizer(stop_words='english', max_features=3000, ngram_range=(1, 2), min_df=3, max_df=0.95), 'text')
], remainder='drop')

# Pipelines
pipelines = {
    'SVM': Pipeline([('pre', preprocessor_svm), ('clf', SVC(kernel='linear', probability=True, class_weight='balanced', random_state=0))]),
    'Random Forest': Pipeline([('pre', preprocessor_rf), ('clf', RandomForestClassifier(n_estimators=100, class_weight='balanced', max_depth=None, min_samples_split=2, max_features='sqrt', random_state=0))]),
    'XGBoost': Pipeline([('pre', preprocessor_xgb), ('clf', XGBClassifier(eval_metric='logloss', scale_pos_weight=3, learning_rate=0.1, random_state=0))]),
    'Logistic Reg.': Pipeline([('pre', preprocessor_lr), ('clf', LogisticRegression(max_iter=1000, class_weight='balanced', random_state=0))]),
    'Multinomial NB': Pipeline([('pre', preprocessor_nb), ('clf', MultinomialNB())])
}
```

```python
records = []
for name, pipe in pipelines.items():
    
    train_data = X_train
    test_data = X_test
        
    pipe.fit(train_data, y_train)
    
    for tag, (data, y_true) in [('Train', (train_data, y_train)), ('Test', (test_data, y_test))]:
        y_pred = pipe.predict(data)
        clf = pipe.steps[-1][1]
        
        rec = {
            'Model': name, 'Dataset': tag,
            'Accuracy': accuracy_score(y_true, y_pred),
            'Precision': precision_score(y_true, y_pred, zero_division=0),
            'Recall': recall_score(y_true, y_pred, zero_division=0),
            'F1-score': f1_score(y_true, y_pred, zero_division=0)
        }
        
        if hasattr(clf, 'predict_proba'):
            rec['ROC-AUC'] = roc_auc_score(y_true, pipe.predict_proba(data)[:, 1])
        else:
            rec['ROC-AUC'] = np.nan
            
        records.append(rec)

df_metrics = pd.DataFrame(records)
```

## Pipeline without metadata features

```python
X = df['text'].astype(str)
y = df['spam'].astype(int)

print(len(X), len(y))  
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0, stratify=y)
```

```python
pipelines = {
    'SVM': Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english', ngram_range=(1, 2), min_df=3, max_df=0.95, max_features=5000)),
        ('clf', SVC(kernel='linear', probability=True, class_weight='balanced', random_state=0))
    ]),
    'Random Forest': Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english', max_features=10000, ngram_range=(1, 2), min_df=5, max_df=0.9)),
        ('clf', RandomForestClassifier(n_estimators=100, class_weight='balanced', max_depth=None, min_samples_split=2, max_features='sqrt', random_state=0))
    ]),
    'XGBoost': Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english', max_features=3000, ngram_range=(1, 2), min_df=3, max_df=0.95)),
        ('clf', XGBClassifier(eval_metric='logloss', scale_pos_weight=3, learning_rate=0.1, random_state=0))
    ]),
    'Logistic Reg.': Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english', ngram_range=(1, 2), min_df=3, max_df=0.95, max_features=5000)),
        ('clf', LogisticRegression(max_iter=1000, class_weight='balanced', random_state=0))
    ]),
    'Multinomial NB': Pipeline([
        ('tfidf', TfidfVectorizer(stop_words='english', max_features=3000, ngram_range=(1, 2), min_df=3, max_df=0.95)),
        ('clf', MultinomialNB())
    ])
}

records = []
for name, pipe in pipelines.items():
    
    pipe.fit(X_train, y_train)
    
    for tag, (X_split, y_true) in [('Train', (X_train, y_train)), ('Test', (X_test, y_test))]:
        y_pred = pipe.predict(X_split)
        clf = pipe.named_steps['clf']
        
        rec = {
            'Model': name, 
            'Dataset': tag,
            'Accuracy': accuracy_score(y_true, y_pred),
            'Precision': precision_score(y_true, y_pred, zero_division=0),
            'Recall': recall_score(y_true, y_pred, zero_division=0),
            'F1-score': f1_score(y_true, y_pred, zero_division=0)
        }
        
        if hasattr(clf, 'predict_proba'):
            rec['ROC-AUC'] = roc_auc_score(y_true, pipe.predict_proba(X_split)[:, 1])
        else:
            rec['ROC-AUC'] = np.nan
            
        records.append(rec)

df_metrics_text_only = pd.DataFrame(records)
```

```python
train_times = {}
test_pred_times = {}

for name, pipe in pipelines.items():
    train_time = timeit.timeit(lambda: pipe.fit(X_train, y_train), number=1)
    train_times[name] = train_time

    predict_fn = lambda: pipe.predict(X_test)
    predict_time = timeit.timeit(predict_fn, number=1)
    test_pred_times[name] = predict_time
times_df = pd.DataFrame({
    'Training Time (s)': train_times,
    'Test Predict Time (s)': test_pred_times})

ax = times_df.plot(kind='bar', figsize=(10, 4))
plt.title('Training Time vs Test Predict Time')
plt.ylabel('Seconds')
plt.xticks(rotation=0)
plt.legend(loc='upper right')
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.3f}s',
                (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

# Project Conclusion

Runtime analysis is a crucial step that separates a standard academic notebook from a truly **production-ready** engineering approach.

---

| Criterion | SVC (Quality Leader) | Logistic Regression (Speed Leader) |
| --- | --- | --- |
| **Metrics (F1 / Accuracy)** | **Maximum** (~99%) | High, but slightly trails behind SVC |
| **Runtime** | **0.63.. sec** (3x slower) | **0.20.. sec** (Lightning-fast) |
| **Scalability** | Will slow down drastically as data grows | Scales beautifully to millions of rows |

---

* **The Victory of Pure NLP over Meta-Features:** Our experiments demonstrated that adding manually engineered features (like email length or exclamation count) actually degraded metrics and confused the models. The best results were achieved when we relied solely on pure text paired with a well-tuned `TfidfVectorizer`.
* **SVC is the Uncompromising Performer:** The Support Vector Classifier excels at separating high-dimensional text embeddings. It is the ideal choice if maximum precision is your only goal, though it demands noticeably more computational power.
* **Logistic Regression is the Ultimate Workhorse:** While losing a tiny fraction of a percent in accuracy, it processes data in just 0.2 seconds. In a real business environment, this massive efficiency boost is often the deciding factor.

---

The ideal model for deployment depends entirely on your specific business requirements:

> **Choose SVC**
> If you are building an anti-spam/phishing filter for confidential corporate emails of high-level executives. A 0.6-second delay per batch is negligible here; what matters most is minimizing the risk of a security breach.

> **Choose Logistic Regression**
> If you are deploying the model for a public email service handling millions of active users, where terabytes of data stream through every second. Logistic Regression will save the company massive amounts of money on server infrastructure while ensuring instantaneous, real-time processing.
